###**Para excesso de velocidade, homens ou mulheres pararam com mais frequência?**

```python
police[police.violation == 'Speeding'].driver_gender.value_counts()
```

###**O gênero afeta quem é revistado durante uma parada?**

```python
police.groupby('driver_gender').search_conducted.sum()
```
